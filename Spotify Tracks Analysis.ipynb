# Spotify track analysis
I will be finding out how music has changed over time. Has the general popularity increased or decreased? what about energy, mode, danceability etc.

```python
# Importing the dataset and libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
```

```python
sns.set_theme(style = 'darkgrid')
music = pd.read_csv("tracks.csv")
music.head()
```

```python
# Checking highest songs based off of popularity
music.sort_values(by='popularity',ascending=False).head()
```

```python
# Checking lowest songs based off of popularity
music.sort_values(by='popularity',ascending=True).head()
```

```python
# Check size of dataset
music.shape
```

Comparing the highest and lowest songs based on popularity, we can already see that the most popular songs are english and have a more recent release date than the least popular songs.

## Cleaning data
First we will remove any duplicates. The id  and id_artists columns will not be used in this analysis which means they'll be dropped. The duration_ms column also needs to be converted into minutes for a better understanding of how long each song is. We will also deal with any null values presented in the dataset. The release date is quite inconsistent so only the year will be accounted for.

```python
music_cleaned = music.drop_duplicates(keep = 'first')
```

```python
music_cleaned.shape
```

No duplicates found

```python
# Deleting unnecessary columns
music_cleaned = music.drop(['id','id_artists'],axis=1)
music_cleaned.head(3)
```

```python
# converting time from ms to min
music_cleaned['duration_min'] = round((music_cleaned['duration_ms'] / 1000) / 60,2)
music_cleaned = music_cleaned.drop(['duration_ms'],axis=1)
music_cleaned.head(3)
```

```python
# Checking any null values in the dataset
pd.isnull(music_cleaned).sum()
```

As we can see, no other column has null values except for the name column. What we can do is replace is the null values with a value like 'N/A' so we could still use it in the analysis.

```python
music_cleaned['name'].fillna('No name', inplace = True)
pd.isnull(music_cleaned).sum()
```

```python
music_cleaned['name'].value_counts().head(10)
```

We can see that No name is an added value in the column with 71 entries.

```python
# Only keep the year in the release_date column, month and day are not relevant
music_cleaned['release_date'] = pd.to_datetime(music_cleaned['release_date'])
music_cleaned['release_date'] =  pd.DatetimeIndex(music_cleaned['release_date']).year
music_cleaned.head(3)
```

All null values are gone, dates are fixed to year, unnecessary columns are removed, data cleaning is finished and ready for analysis.

## Analyzing data
We can now finally analyze how the overall state of music changed over time.

Before anything, let's see the correlation between each factor in the dataset

```python
plt.figure(figsize=(10,10))
sns.heatmap(music_cleaned.corr(), annot = True, fmt = ".2f")
plt.show()
```

The closer the absolute value of the number is closer to 1, the stronger the correlation is between the two variables. When we look at the release date column, There are a couple of variables which have medium - strong correlation such as popularity, energy, loudness, acousticness etc.)

Let's first visualize popularity of song over time

```python
# Scatter plot to see individual song populatity over time
sns.set_palette("bright")
sns.scatterplot(x = 'release_date', y = 'popularity', data = music_cleaned)
plt.xlabel('Year of Release')
plt.ylabel('Popularity')
plt.title('Popularity of music over time')
plt.show()
```

```python
# Linegraph to see average popularity over year
sns.set_palette("bright")
sns.lineplot(x = 'release_date', y = 'popularity', data = music_cleaned)
plt.xlabel('Year of Release')
plt.ylabel('Popularity')
plt.title('Popularity of music over time')
plt.show()
```

Looking at the average popularity of songs over time, we notice that from 1900 - 1920, there is a linear dip to almost zero. This is because there is only one song before 1920. Then from 1920 - 1950 the popularity stays consistent at around 0. Then from 1950 - 1990, we see that the overall popularity of music is slowly increasing. From 1990 onwards, there is a dramatic increase in the popularity of music (NOTE: slight dip around 2021 signifies that new releases have not caught up to the streams of older releases).

Looking at the popularity of songs over time allows us to question what are the reasons the popularity dramatically changed? By analyzing other factors we can find out what has the largest effect on music popularity over time.

## Determining change: Old vs New
to determine how other how songs changed over time, we will break down the data into four sections. These sections were determined by the popularity distribution in the above line graph.

low_music: Music from 1900 - 1950

mid_music: Music from 1950 - 1990

new_music: Music from 1990 onwards

```python
# Split data into four sections
early_music = music_cleaned.loc[music_cleaned['release_date'] < 1950]
mid_music = music_cleaned.loc[(music_cleaned['release_date'] >= 1950) & (music_cleaned['release_date'] < 1990)]
new_music = music_cleaned.loc[music_cleaned['release_date'] >= 1990]
```

```python
early_music['popularity'].describe()
```

```python
mid_music['popularity'].describe()
```

```python
new_music['popularity'].describe()
```

We can see that there is only 1 song before 1920, which is why there is a linear decrease in the graph. The average popularity slowly increases over time, it jumps from 1 to 20 to 35. One thing to note is that spotify was founded in 2006 so older entries of songs may not be represented as fairly since the streams prior to 2006 could not have been counted.

Let's now compare the energy of older songs vs newer songs, and see how much change there is

```python
plt.figure(figsize=(17, 10))
# Create histogram of early music
plt.subplot(2, 2, 1)
plt.hist(early_music['energy'], range = (0,1), color = 'orange')
plt.title('Energy of music before 1950')
plt.xlabel('Energy')
plt.ylabel('Count')

# Create histogram of mid_music
plt.subplot(2,2,2)
plt.hist(mid_music['energy'], color = 'green', range = (0,1))
plt.title('Energy of music after 1950 and before 1990')
plt.xlabel('Energy')
plt.ylabel('Count')

# Create histogram of new_music
plt.subplot(2,2,3)
plt.hist(new_music['energy'], color = 'red', range = (0,1))
plt.title('Energy of music after 1990')
plt.xlabel('Energy')
plt.ylabel('Count')

# Create collective histogram
plt.subplot(2,2,4)
plt.hist(new_music['energy'], color = 'red', range = (0,1), label = '1990+')
plt.hist(mid_music['energy'], color = 'green', range = (0,1), label = '1950-1990')
plt.hist(early_music['energy'], range = (0,1), label = '0-1920', color = 'orange')
plt.legend()
plt.title('Comparison of Energy between time periods')
plt.xlabel('Energy')
plt.ylabel('Count')
plt.show()
```

```python
# Let's check average duration
print(early_music['energy'].mean())
print(mid_music['energy'].mean())
print(new_music['energy'].mean())
```

With some dramatic findings, we can see that the energy of songs over the three major time periods goes from right skewed, to left skew signifying that the energy of songs on average has increased over time.

## Energy and loudness

In correlation to release date, we can see that loudness and energy have almost the same correlation value (0.45 and 0.46), and they have a correlation value of 0.76. We can visualize this observation. We will remove the only entry from before 1920 so the graph looks neater.

```python
music_cleaned = music_cleaned.drop(music_cleaned[music_cleaned['release_date'] == 1900].index, axis=0)

plt.figure(figsize=(17, 10))

plt.subplot(2, 2, 1)
sns.lineplot(x = 'release_date', y = 'energy', data = music_cleaned)

plt.xlabel('Release Date')
plt.ylabel('Energy')
plt.title('Release Date vs. Energy')

plt.subplot(2, 2, 2)
sns.lineplot(x = 'release_date', y = 'loudness', data = music_cleaned)

plt.xlabel('Release Date')
plt.ylabel('Loudness')
plt.title('Release Date vs. Loudness')

plt.subplot(2, 2, 3)
sns.scatterplot(x = 'energy', y = 'loudness', data = music_cleaned)

plt.xlabel('Energy')
plt.ylabel('Loudness')
plt.title('Energy vs. Loudness')

plt.show()
```

We can see that there is a relation of whenever a song is louder, the more energy is has. This can be seen by the positive correlation in the Energy vs. Loudness scatter plot and how similarly energy progressed over time compared to loudness.

## Acousticness

So far, we've seen popularity, energy, and loudness increase over time. On the correlation heat map it can bee seen that the acousticness of songs have generally decreased over time.

It can be also notified that there is a direct relation of acousticness and energy. Let's analyze this further.

```python
plt.figure(figsize=(17, 10))

plt.subplot(1, 2, 1)
sns.lineplot(x = 'release_date', y = 'acousticness', data = music_cleaned)

plt.xlabel('Release Date')
plt.ylabel('Acousticness')
plt.title('Release Date vs. Acousticness')


plt.subplot(1, 2, 2)
sns.lineplot(x = 'energy', y = 'acousticness', data = music_cleaned)

plt.xlabel('Energy')
plt.ylabel('Acousticness')
plt.title('Energy vs. Acousticness')

plt.show()
```

It is evident that over time, the acousticness of songs have declined, this is possibly due to the increase of energy and loudness in songs. This is also possibly due to technology becoming more advanced and the rise of electronic instruments. These electronic instruments also amplify the energy and loudness of a song.

## Relevancy of time periods

There are certain time periods where society as a whole may not be in the greatest position, such as war or economic problems. These situations can be translated through music, are songs sadder during those times? Happier? Let's see if there are any findings.

We will analyze music from 1929 - 1945 where the Great Depression and World War 2 occured. We will compare the musical elements in this period to a more recent time period of 2005 - 2020.

```python
ww2 = music_cleaned.loc[(music_cleaned['release_date'] >= 1929) & (music_cleaned['release_date'] <= 1944)]
recent = music_cleaned.loc[(music_cleaned['release_date'] >= 2005) & (music_cleaned['release_date'] <= 2020)]
```

Let's first compare the modes of songs in each time period (0 = minor (sad) and 1 = major (happy))

```python
ww2_counts = ww2['mode'].value_counts()
rec_counts = recent['mode'].value_counts()

plt.figure(figsize=(17, 10))

plt.subplot(2,2,1)
sns.barplot(x=rec_counts.index, y=rec_counts.values, color='orange')
plt.xlabel('Mode')
plt.ylabel('Count')
plt.title('The mode of recent songs')

plt.subplot(2,2,2)
sns.barplot(x=ww2_counts.index, y=ww2_counts.values, color='blue')
plt.xlabel('Mode')
plt.ylabel('Count')
plt.title('The mode of older songs')

plt.subplot(2,2,3)
plt.hist(recent['tempo'], range = (0,250), label = '0-1920', color = 'orange')
plt.xlabel('Tempo (BPM)')
plt.ylabel('Count')
plt.title('The tempo of recent songs')

plt.subplot(2,2,4)
plt.hist(ww2['tempo'], range = (0,250), label = '0-1920', color = 'blue')
plt.xlabel('Tempo (BPM)')
plt.ylabel('Count')
plt.title('The tempo of older songs')


plt.show()
```

Surprisingly, there were no significant differences found. One thing to note is that there are a higher amount of minor songs proportionally in the recent era rather than the older era. Songs that are recent are slightly but not significantly faster (higher tempo).

## Conclusion

In conclusion, we have found out many things which changed in music over the years. Popularity has significantly increased, noting that this information is based on a streaming platform founded in 2006. 

The energy and loudness of music have dramatically increased over time, which also caused a decrease in acoustic instruments in songs. The use of technology and electronic instruments also may have caused an a decrease in acousticness.

Comparing two very different periods in time, we have found that the distribution of modes in music do not change depending on time periods, although the tempo has slightly changed when comparing the two periods.
